# Classification

###  import needed library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn import tree
```

###  

### print the head of data 

```python
df=pd.read_csv("selected_data_with_target.csv")
print(df.head(5))
```

###  

### split dataset in features and target variable

```python
fn=df.keys().tolist()[:-1]
x=df[fn]
y=df['stroke']
print(x)
```

### 

### Gini index (25% test, 75% training)

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
cn=df['stroke'].unique()
disp=ConfusionMatrixDisplay.from_estimator(clf,x_test,y_test,display_labels=cn)
```

```python
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(clf, 
               feature_names=fn,
               class_names=[str(cn[i]) for i in range(len(cn))],
               filled=True);
```

### 

### IG (entropy) (25% test, 75% training)

```python
clf=DecisionTreeClassifier(criterion="entropy")
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
cn=df['stroke'].unique()
disp=ConfusionMatrixDisplay.from_estimator(clf,x_test,y_test,display_labels=cn)
```

```python
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(clf, 
               feature_names=fn,
               class_names=[str(cn[i]) for i in range(len(cn))],
               filled=True);
```
