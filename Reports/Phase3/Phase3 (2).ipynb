# Classification

###  import needed library

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn import tree
```

### print the head of data 

```python
df=pd.read_csv("selected_data_with_target.csv")
print(df.head(5))
```

###  

### split dataset in features and target variable

```python
fn=df.keys().tolist()[:-1]
x=df[fn]
y=df['stroke']
print(x)
```

### 

### Gini index (30% test, 70% training)

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
cn=df['stroke'].unique()
disp=ConfusionMatrixDisplay.from_estimator(clf,x_test,y_test,display_labels=cn)
```

```python
fig, axes=plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(clf, 
               feature_names=fn,
               class_names=[str(cn[i]) for i in range(len(cn))],
               filled=True);
```

# Clustering
The clustering process entails the grouping together of similar data points based on their shared characteristics or similarities. One widely used clustering technique is k-means, which divides the dataset into k clusters, assigning each data point to the cluster with the closest mean value determined by Euclidean distance.

By applying the k-means algorithm to the Stroke Prediction Dataset, after excluding the stroke class label and converting attributes to numerical format, we can effectively conduct clustering analysis to detect patterns and groupings within the data. This approach helps uncover valuable insights and potentially reveals hidden relationships among the variables, thus enriching our comprehension of stroke risk factors and aiding in the development of improved healthcare strategies.

To implement clustering, we can follow several steps. Initially, we eliminate the "stroke" class label from the dataset since clustering is an unsupervised learning task and does not necessitate labeled data.







```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import numpy as np 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import scale
df = pd.read_csv("selected_data_with_target.csv")
print(df.head())
```

```python
features = df.drop(df.columns[-1], axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)
print("\nScaled DataFrame:")
print(df_scaled.head())
```

We can proceed by implementing the k-means clustering algorithm. The first step is to choose a suitable value for k, which signifies the number of clusters we aim to detect within the data. To ascertain the optimal k value, we experiment with various sizes such as 2, 4, or 5, and then evaluate the clustering outcomes for each.

After executing the k-means algorithm with different k values, we evaluate the clustering quality using diverse metrics such as Silhouette coefficient, Elbow (total within-cluster sum of square). These metrics aid us in identifying the most appropriate number of clusters that yield the most meaningful and distinct groups.

The average silhouette score quantifies how closely an object aligns with its own cluster in comparison to neighboring clusters, ranging from -1 to 1. A higher score indicates better alignment with its own cluster and poorer alignment with neighboring ones.

The Within-Cluster Sum of Squares (WSS) measures the compactness of clusters, providing insights into how tightly the data points are clustered within each group.

In determining the optimal k value, we experiment with different sizes, such as 2, 4, or 5.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.cluster import KMeans
#5) Build the K-means model
# Set a seed for random number generation
np.random.seed(8953)

# Define different values of K
k_values = [2, 4, 5]  # Example: using 2, 4, and 5 clusters

# Iterate over each value of K
for k in k_values:
    # Run K-means clustering with the current value of K
    kmeans = KMeans(n_clusters=k, random_state=8953)
    kmeans_result = kmeans.fit(df_scaled)

    # Print the clustering result for the current value of K
    print(f"Cluster Centers (K={k}):")
    print(kmeans_result.cluster_centers_)
    print("\nCluster Labels:")
    print(kmeans_result.labels_)
    print("\n")
```

The second evaluation method employed is the Elbow method, which identifies the optimal number of clusters as the point where the curve exhibits a significant change in direction.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming df_scaled is already defined and contains your scaled data

# Calculate the total within-cluster sum of squares (WCSS) for different values of k
wcss = []
for i in range(1, 11):  # Try different values of k from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=8953)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Using the elbow method we can see that the best number of clusters is k=5.

from our dive in clustering we can see that the best number of clusters for our data is k=5.
