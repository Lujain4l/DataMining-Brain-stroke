### Imbalanced dataset problem:
Our dataset initially suffered from an imbalance issue, meaning there were far more instances of one class (no stroke) than the other (stroke). This can skew the results of data analysis and modeling.

To address this, we randomly removed some samples from the majority class (no stroke) until we achieved a more balanced distribution, aiming for a 40%-60% split between the two classes. This rebalancing ensures that our model learns from a fair representation of both outcomes, improving its ability to make accurate predictions for strokes.

```python



# Specify the desired number of non-stroked samples after balancing
desired_non_stroked = 260

# Calculate the number of non-stroked samples to drop
rows_to_drop = len(df[df['stroke'] == 0]) - desired_non_stroked

# Drop rows randomly from the non-stroked class
if rows_to_drop > 0:
    df_balanced = df.drop(df[df['stroke'] == 0].sample(n=rows_to_drop, random_state=42).index)
else:
    df_balanced = df.copy()

# Print the class distribution in the balanced dataset
print(df_balanced['stroke'].value_counts())
# If you want to save the balanced dataset into a new DataFrame, you can do:
balanced_df = df_balanced.copy()
print(balanced_df)
# Now, df_balanced contains the balanced dataset with the specified number of non-stroked samples.
df=balanced_df
df.to_csv("balancedData.csv")
```

### 5.1 Classification

We used the decision tree approach, which builds a tree iteratively with leaf nodes indicating the ultimate decisions. Our goal was to predict the class label "stroke," which is divided into two categories: "yes" and "no." The prediction was "age", "avg_glucose_level", "hypertension", and "heart_disease" which were obtained from the feature selection results.

The classification technique involved splitting the dataset into two subsets: 
Training dataset: Used for constructing the decision tree model.
Testing dataset: Employed to assess the performance of the constructed model.

Also, we have used three different partitions to determine which model will predict well. We tried three different sizes of partitions and two attribute selection measures: Information gain (IG) and Gini index. 

Information gain:
Information gain and entropy are used to measure the usefulness of a feature in a dataset. 
Information gain measures how much information a particular feature contributes to the classification process. It quantifies the reduction in entropy achieved by splitting the data based on that feature.
Entropy is a measure of the impurity or disorder within a set of data. In the context of decision trees, it quantifies the randomness or uncertainty in the target variable.

To use information gain in a decision tree, the algorithm examines different features and calculates the information gain for each one. The feature with the highest information gain is selected as the best choice for splitting the data.
The decision tree algorithm seeks to produce subsets that are more homogeneous or pure in terms of the class labels by choosing features with high information gain. 


Gini index:
The Gini index is a metric used in decision tree algorithms for measuring impurity or the degree of disorder within a set of data. In decision tree algorithms, the Gini index is used to evaluate the quality of potential splits at each node. The split that minimizes the Gini index is selected as the optimal splitting criterion, as it leads to the maximum possible purity or homogeneity in the resulting subsets.

### 5.2 Clustering

The clustering process entails the grouping together of similar data points based on their shared characteristics or similarities. One widely used clustering technique is k-means, which divides the dataset into k clusters, assigning each data point to the cluster with the closest mean value determined by Euclidean distance.

By applying the K-means algorithm to the Stroke Prediction Dataset, after excluding the stroke class label and converting attributes to numerical format, we can effectively conduct clustering analysis to detect patterns and groupings within the data. This approach helps uncover valuable insights and potentially reveals hidden relationships among the variables, thus enriching our comprehension of stroke risk factors and aiding in the development of improved healthcare strategies.

For clustering analysis on the dataset, we,ve utilized the K-means clustering technique. Here's a breakdown of the process, along with the rationale and implementation details:

1. **Data Preprocessing:**
   - Initially, I imported the required libraries, including pandas, numpy, seaborn, matplotlib.pyplot, and sklearn. I also loaded the dataset into a pandas DataFrame and standardized the features using StandardScaler from sklearn.preprocessing.

2. **K-means Clustering:**
   - K-means clustering is a popular unsupervised learning algorithm used for partitioning data into K clusters based on feature similarity.
   - In Python, I used the KMeans class from sklearn.cluster to implement K-means clustering. This method allows us to specify the number of clusters (K) and other parameters like random_state for reproducibility.
   - I defined different values of K (e.g., 2, 4, 5) to explore various cluster configurations.
   - For each value of K, I ran the K-means algorithm and obtained cluster centers and labels.
   - The resulting cluster labels were added to the original DataFrame to facilitate further analysis and visualization.

3. **Visualization:**
   - To visualize the clustering results, I used matplotlib.pyplot and seaborn libraries to create scatter plots.
   - Each scatter plot represents the data points colored by their assigned cluster labels. This allows for a visual inspection of how well the clustering algorithm segregates the data into distinct groups.
   - Additionally, I employed the SilhouetteVisualizer from the yellowbrick.cluster module to visualize the silhouette scores, which measure the separation distance between clusters.

4. **Elbow Method:**
   - Lastly, I implemented the Elbow Method to determine the optimal number of clusters (K). This method involves plotting the within-cluster sum of squares (WCSS) against the number of clusters and identifying the "elbow point," which signifies the point of diminishing returns in terms of clustering improvement.

Overall, these techniques provide a comprehensive approach to clustering analysis, enabling data exploration, visualization, and determination of optimal cluster configurations.

### 6.2 Clustering

To implement clustering, we can follow several steps. Initially, we eliminate the "stroke" class label from the dataset since clustering is an unsupervised learning task and does not necessitate labeled data.

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import numpy as np 
from sklearn.cluster import KMeans 
from sklearn.preprocessing import scale
df = pd.read_csv("selected_data_with_target.csv")
print(df.head())
```

```python
features = df.drop(df.columns[-1], axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)
print("\nScaled DataFrame:")
print(df_scaled.head())
```

We can proceed by implementing the k-means clustering algorithm. The first step is to choose a suitable value for k, which signifies the number of clusters we aim to detect within the data. To ascertain the optimal k value, we experiment with various sizes such as 2, 4, or 5, and then evaluate the clustering outcomes for each.

After executing the k-means algorithm with different k values, we evaluate the clustering quality using diverse metrics such as Silhouette coefficient, Elbow (total within-cluster sum of square). These metrics aid us in identifying the most appropriate number of clusters that yield the most meaningful and distinct groups.

The average silhouette score quantifies how closely an object aligns with its own cluster in comparison to neighboring clusters, ranging from -1 to 1. A higher score indicates better alignment with its own cluster and poorer alignment with neighboring ones.

The Within-Cluster Sum of Squares (WSS) measures the compactness of clusters, providing insights into how tightly the data points are clustered within each group.

In determining the optimal k value, we experiment with different sizes, such as 2, 4, or 5.

```python
%pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
#5) Build the K-means model
# Set a seed for random number generation
np.random.seed(8953)

# Define different values of K
k_values = [2, 4, 5]  # Example: using 2, 4, and 5 clusters

# Iterate over each value of K
for k in k_values:
    # Run K-means clustering with the current value of K
    kmeans = KMeans(n_clusters=k, random_state=8953)
    kmeans_result = kmeans.fit(df_scaled)

    # Print the clustering result for the current value of K
    print(f"Cluster Centers (K={k}):")
    print(kmeans_result.cluster_centers_)
    print("\nCluster Labels:")
    print(kmeans_result.labels_)
    print("\n")

    # Add cluster labels to the original DataFrame
    df_scaled['Cluster'] = pd.Series(kmeans_result.labels_)

    # Visualize clustering using a scatter plot
    # We used 'Murder' and 'Assault' as features (variables) for plotting
    plt.figure(figsize=(10, 6))
    sns.scatterplot (x=df_scaled.columns [0], y=df_scaled.columns [1], hue='Cluster', data=df_scaled, palette='viridis', edgecolor='k')
    plt.title(f'K-Means Clustering Results for K={k}')
    plt.xlabel(df_scaled.columns[0])
    plt.ylabel(df_scaled.columns [1])
    plt.legend()
    plt.show()
```

The second evaluation method employed is the Elbow method, which identifies the optimal number of clusters as the point where the curve exhibits a significant change in direction.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming df_scaled is already defined and contains your scaled data

# Calculate the total within-cluster sum of squares (WCSS) for different values of k
wcss = []
for i in range(1, 11):  # Try different values of k from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=8953)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
